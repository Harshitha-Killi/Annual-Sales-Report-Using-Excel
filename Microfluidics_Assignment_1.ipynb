Non-dimensional Velocity Profile for Couette Flow with Slip with constant σv and varying Kn:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values for Kn and σv
Kn_values = [0.115, 0.1, 0.05, 0.01]
σv_values = [0.8]

# Create a figure for the plot
plt.figure(figsize=(8, 6))

# Iterate through combinations of Kn and σv
for Kn in Kn_values:
    for σv in σv_values:
        # Generate y values within a reasonable range
        y_values = np.linspace(0, 1, 400)

        # Calculate u values using the equation
        U_values = (y_values + (2 - σv) * Kn / σv)/(1 + 2 * Kn * (2 - σv) / σv)

        # Plot the line
        plt.plot(U_values, y_values, label=f'Kn={Kn:.3f}, σv={σv:.3f}')

# Customize the plot
plt.title('Non-dimensional Velocity Profile for Couette Flow with Slip with constant σv and varying Kn')
plt.xlabel('U/U∞ (Non-dimensional velocity)')
plt.ylabel('y/H (Non-dimensional distance from stationary plate)')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

Non-dimensional Velocity Profile for Couette Flow with Slip with constant Kn and varying σv:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values for Kn and σv
Kn_values = [0.15]
σv_values = [1.0,0.9,0.8,0.7]

# Create a figure for the plot
plt.figure(figsize=(8, 6))

# Iterate through combinations of Kn and σv
for Kn in Kn_values:
    for σv in σv_values:
        # Generate y values within a reasonable range
        y_values = np.linspace(0, 1, 400)

        # Calculate u values using the equation
        U_values = (y_values + (2 - σv) * Kn / σv)/(1 + 2 * Kn * (2 - σv) / σv)

        # Plot the line
        plt.plot(U_values, y_values, label=f'Kn={Kn:.3f}, σv={σv:.3f}')

# Customize the plot
plt.title('Non-dimensional Velocity Profile for Couette Flow with Slip with constant Kn and varying σv')
plt.xlabel('U/U∞ (Non-dimensional velocity)')
plt.ylabel('y/H (Non-dimensional distance from stationary plate)')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

Non-dimensional Velocity Profile for Couette Flow without Slip:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of values for x and y
x_range = np.linspace(0, 1, 100)  # Adjust the range and number of points as needed

# Calculate y values based on the equation x = y
y_values = x_range

# Create a figure for the plot
plt.figure(figsize=(8, 6))

# Plot the line x = y
plt.plot(x_range, y_values, label='U/U∞ = y/H')

# Customize the plot
plt.title('Non-dimensional Velocity Profile for Couette Flow without Slip')
plt.xlabel('U/U∞ (Non-dimensional velocity)')
plt.ylabel('y/H (Non-dimensional distance from stationary plate)')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

**Comparision between the Slip and No-Slip conditions:**

Slip Condition:
*   The velocity profile does not pass through the origin.
*   There is an intercept on the x-axis.
*   The profile represents ax+by+c=0 equation.
*   Velocity profile is a function of Kn and σv.



No-Slip Condition:
*   Velocity profile passes through the origin.
*   There is no intercept on the x-axis.
*   The profile represents x=y equation.
*   Velocity profile is not a function of Kn and σv.
